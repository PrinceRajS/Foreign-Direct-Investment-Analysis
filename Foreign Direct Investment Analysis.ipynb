## Data Loading

```python
import pandas as pd

df = pd.read_csv('FDI data.csv')

# Displaying the first 10 rows of the dataset
df.head(10)
```

## Exploratory Data Analysis (EDA)

```python
df.describe()
```

### Sector-wise Analysis

```python
# Sector-wise total investment
sector_investment = df.groupby('Sector').sum()
sector_investment.head(100)
```

```python
# Plotting sector-wise investment
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(30, 18))
sector_investment.sum(axis=1).sort_values().plot(kind='barh')
plt.title('Total FDI by Sector')
plt.xlabel('Total FDI (in millions)')
plt.ylabel('Sector')
plt.show()
```

### Year-wise Analysis

```python
# Assuming the first column is 'Sector'
year_columns = df.columns[1:]
df[year_columns] = df[year_columns].apply(pd.to_numeric, errors='coerce')
```

```python
# Year-wise total investment
year_investment = df[year_columns].sum()
print(year_investment)
```

## Key Metrics and Relationships

```python
# Calculating correlation matrix
correlation_matrix = df[year_columns].corr()
print(correlation_matrix)
```

### Correlation Analysis

```python
# Visualizing the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Calculating annual growth rate
growth_rate = year_investment.pct_change().fillna(0) * 100
print(growth_rate)
```

```python
# Plotting growth rate
plt.figure(figsize=(10, 6))
growth_rate.plot(kind='bar')
plt.title('Annual Growth Rate of FDI')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.show()
```

### Sector-wise Detailed Analysis

```python
# Identify top sectors
top_sectors = sector_investment.sum(axis=1).sort_values(ascending=False).head(5)
print(top_sectors)
```

```python
# Plot top sectors over the years
plt.figure(figsize=(12, 8))
for sector in top_sectors.index:
    plt.plot(year_columns, df[df['Sector'] == sector].iloc[0, 1:], label=sector)

plt.title('FDI Trends in Top Sectors')
plt.xlabel('Year')
plt.ylabel('FDI (in millions)')
plt.legend()
plt.show()
```

### Sector-wise Contribution to Total FDI

```python
# Calculate sector-wise contribution
sector_contribution = (sector_investment.sum(axis=1) / sector_investment.sum(axis=1).sum()) * 100
print(sector_contribution)
```

```python
# Plot sector-wise contribution
plt.figure(figsize=(60, 40))
sector_contribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title('Sector-wise Contribution to Total FDI')
plt.ylabel('')
plt.show()
```

### Seasonal Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Performing seasonal decomposition
decomposition = seasonal_decompose(year_investment, model='additive', period=1)
decomposition.plot()
plt.show()
```
